```python
from collections import OrderedDict

lbtt_bands = OrderedDict([
    (0, 0),
    (145000, 0.02),
    (250000, 0.05),
    (325000, 0.1),
    (750000, 0.12),
])


def get_lbtt_amount(value):
    lbtt_value = 0
    for thresh, rate in reversed(lbtt_bands.items()):
        applicable_value = max(0, value - thresh)
        lbtt_value += applicable_value * rate
        value -= applicable_value
    return lbtt_value
```

```python
funds = 50000
home_report = 280000
```

```python
m_deposit = 0.1 * home_report
m_deposit
```

```python
available_funds = funds - m_deposit
available_funds
```

```python
maximum_offer = home_report + available_funds
maximum_offer
```

```python
maximum_offer_less_lbtt = maximum_offer - get_lbtt_amount(maximum_offer)
maximum_offer_less_lbtt
```

```python
maximum_offer_less_lbtt + get_lbtt_amount(maximum_offer_less_lbtt)
```
